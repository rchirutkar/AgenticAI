
```

Coding Agent:   You describe goal → agent plans → writes → tests → iterates

```

**The loop:**
```
Describe task
    → Agent creates a plan
    → Writes code across multiple files

```

**When coding agents are most useful:**
- Boilerplate-heavy tasks (CRUD, API integrations)
- Refactoring across many files
- Writing tests for existing code
- Scaffolding a new project
